```python
def remove_punctuation(text):
  import string
  translator = str.maketrans('','',string.punctuation)
  return text.translate(translator)
```

```python
popular_df['Serial_overview'] = popular_df['Serial_overview'].apply(remove_punctuation)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
np.array(sw)
```

```python
#stop_words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
```

```python
'''a function for removing the stopword'''
```

```python
def stopwords(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
popular_df['Serial_overview'][0]
```

```python
popular_df['Serial_overview'] = popular_df['Serial_overview'].apply(stopwords)
```

```python
from nltk.stem.snowball import SnowballStemmer
snowball = SnowballStemmer("english")
```

```python
stemmer = SnowballStemmer("english")

def stemming(text):    
    '''a function which stems each word in the given text'''
    text = [stemmer.stem(word) for word in text]
    return"" "".join(text) 
```

```python
popular_df['Serial_overview'] = popular_df['Serial_overview'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def tfidf_vectorizer(text):
  tfidf_vectorizer_func = TfidfVectorizer()
  tfidf_vectorizer_func.fit(popular_df['Serial_overview'])
  dictionary = tfid_vectorizer_func.vocabulary_.items()
  return dictionary
```

```python
vectorizer = TfidfVectorizer(sublinear_tf= True, min_df=10, norm='l2', ngram_range=(1, 2), stop_words='english')
X_vc = vectorizer.fit_transform(popular_df["Serial_overview"])

pt = pd.DataFrame(X_vc.toarray(), columns=vectorizer.get_feature_names_out()).head(724)
```

```python
def recommend(serial_name):
    serial_index = popular_df[popular_df['Name'] == serial_name].index[0]
    distances = similarity_score[serial_index]
    serial_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in serial_list:
        print(popular_df.iloc[i[0]])
   
```
